```python
#Filtering Ordering

df2 = pd.read_csv('/content/world_population.csv')
df2
```

```python
df2[df2['Country']=='Greece'] # Find specific  country's name
```

```python
gr_ger = ['Greece','Germany'] # I specify these countries in a list and try to find if include in df2 with isin() function

df2[df2['Country'].isin(gr_ger)]
```

```python
df2[df2['Country'].str.contains('United')]
```

```python
df2[df2['Country'].str.contains('To')]
```

```python
df3 = df2.set_index('Country')
df3
```

```python
df3.filter(items = ['CCA3','Continent']) # axes can be 0 or 1 horizontal or defeault
```

```python
df3.filter(items = ['Greece'],axis=0)
```
